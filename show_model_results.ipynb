```python
import json
import os
import pandas as pd
import pickle
import re
```

```python
trained_model_path = './trained_models/'
model_ids = os.listdir(trained_model_path)
models = [f for f in model_ids if 'model' in f]
ids = [re.search(r'model_(.*?).pt', m).group(1) for m in models]
ids
```

```python
results = []
for model_id in ids:
    params_file = f'{trained_model_path}/params_{model_id}.json'
    with open(params_file, 'r') as f:
        params = json.load(f)
    metrics_file = f'{trained_model_path}/metrics_{model_id}.pickle'
    with open(metrics_file, 'rb') as f:
        metrics = pickle.load(f)
    model_results = {'model_id': model_id, **params, **metrics}
    results.append(model_results)

pd.DataFrame(results).sort_values(by='mIoU', ascending=False)    
```
